## Summary

| Algorithm | Strengths | Weaknesses | Key Takeaway |
|-----------|-----------|------------|--------------|
| **k-Means** | Fast, simple, vectorized | Assumes spherical clusters | Fails on anisotropic data |
| **Gaussian NB** | Few parameters, stable | Independence assumption | Overconfident on correlated features; fails on XOR |
| **C4.5 Tree** | Interpretable, axis-aligned | Greedy, high variance | Prone to overfitting; cannot learn parity |
